```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/drive/MyDrive/Enigma Genomics/PGS001298_hmPOS_GRCh38.txt"
```

```python
df = pd.read_csv(file_path,sep="\t",comment="#")
print("Shape:", df.shape)
print(df.head())
print(df.columns)
print("Data types:")
print(df.dtypes)
print(df.isnull().sum())
```

```python
chr21_df = df[(df['hm_chr'] == '21') & (df['hm_pos'].notna())]
print("Chromosome 21 shape:", chr21_df.shape)
display(chr21_df[['hm_chr', 'hm_pos', 'effect_weight']].head())
print("\nSummary statistics:")
display(chr21_df['effect_weight'].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(chr21_df['effect_weight'], bins=30, kde=True)
plt.title("Distribution of Effect Weight â€“ Chromosome 21 (PGS001298, HmPOS GRCh38)")
plt.xlabel("Effect Weight")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
